Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Play Store dataset
url = "https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv"
playstore_data = pd.read_csv(url)

# Filter out rows with missing values in 'Type' or 'Price' columns
playstore_data = playstore_data.dropna(subset=['Type', 'Price'])

# Convert 'Price' column to numeric by removing '$' and converting to float
playstore_data['Price'] = playstore_data['Price'].apply(lambda x: 0 if x == 'Everyone' else float(x.replace('$', '')))

# Calculate average prices by app type
average_prices = playstore_data.groupby('Type')['Price'].mean().reset_index()

# Create a bar chart to compare average prices by app type
plt.figure(figsize=(10, 6))
plt.bar(average_prices['Type'], average_prices['Price'], color='skyblue')
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

```python
import pandas as pd

# Load the Google Play Store dataset
url = "https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv"
playstore_data = pd.read_csv(url)

# Remove rows with missing values in the 'Installs' column
playstore_data = playstore_data.dropna(subset=['Installs'])

# Remove special characters from the 'Installs' column and convert to integer
playstore_data['Installs'] = playstore_data['Installs'].str.replace('[+,]', '').replace('Free', '0').astype(int)

# Group by app name and calculate the total number of installs
install_counts = playstore_data.groupby('App')['Installs'].sum().reset_index()

# Sort the install counts in descending order
install_counts_sorted = install_counts.sort_values(by='Installs', ascending=False)

# Get the top 10 most popular apps
top_10_apps = install_counts_sorted.head(10)

print("Top 10 most popular apps based on number of installs:")
print(top_10_apps)
```

Q17. A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to
launch a new app. What features would you analyze to make recommendations to the company, and
what data visualizations would you use to present your findings?

To identify the best time to launch a new app on the Google Play Store, several features in the dataset can be analyzed to understand seasonal trends, user behavior, and market dynamics. Heres a comprehensive approach:

Key Features to Analyze:
Last Updated: Analyze the 'Last Updated' date of apps to understand when successful apps were launched or updated. This can help identify any seasonal patterns or trends in app popularity.

Installs Over Time: Look at the number of installs over different time periods (e.g., months, quarters) to identify peak times when users are more likely to download new apps.

Ratings Over Time: Examine the average ratings and user reviews over different time periods to understand user satisfaction and engagement during different seasons.

App Categories: Analyze the distribution of installs and ratings across different app categories over time. Some categories may have seasonal popularity spikes.

Promotional Periods: Identify periods that correspond to known promotional events (e.g., holidays, Black Friday, Cyber Monday) to see if there is an increase in installs and user activity during these times.

Data Visualizations to Use:
Time Series Analysis:

Line Charts: Use line charts to visualize the number of installs over time for different app categories. This helps in identifying trends and peak periods.
Seasonal Decomposition: Apply seasonal decomposition of time series (e.g., using statsmodels) to separate and analyze the seasonal component of the install data.
Bar Charts and Histograms:

Monthly/Quarterly Installs: Use bar charts to compare the average number of installs per month or quarter. This can help in identifying specific months or quarters with higher user activity.
Install Frequency by Launch Date: Create histograms to show the frequency of app launches by month or quarter, highlighting popular launch periods.
Heatmaps:

Monthly Installs Heatmap: Use heatmaps to visualize the number of installs per month across different years. This can reveal any consistent seasonal patterns.
Category-Season Heatmap: Create a heatmap to show the popularity of different app categories during different seasons (e.g., winter, spring, summer, fall).

Installs Distribution by Month: Use box plots to show the distribution of installs for each month. This can help in understanding the variability and identifying months with consistently high installs.
Calendar Heatmaps:

Daily Installs: Use calendar heatmaps to visualize the number of installs on a daily basis. This can help identify specific days or weeks with unusually high or low activity.
Steps to Perform the Analysis:
Data Preparation: Clean the dataset by handling missing values, converting date columns to datetime format, and ensuring the data is in a usable format.

Extract Time Features: Extract relevant time-based features (e.g., month, quarter, year) from the 'Last Updated' column.

Aggregating Data: Aggregate the data by time periods (e.g., monthly, quarterly) to calculate the total and average number of installs, average ratings, and number of reviews.

Visualize and Interpret: Create the visualizations mentioned above and interpret the results to identify the best time periods for launching a new app.

Report Findings: Summarize the findings in a report with visual aids, highlighting key trends and recommendations for the best time to launch a new app.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Google Play Store dataset
url = "https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv"
playstore_data = pd.read_csv(url)

# Remove rows with missing values in the 'Last Updated' column
playstore_data = playstore_data.dropna(subset=['Last Updated'])

# Function to check and correct invalid dates
def correct_invalid_dates(date_str):
    try:
        return pd.to_datetime(date_str)
    except ValueError:
        return pd.NaT  # Not a Time (missing value)

# Apply the function to the 'Last Updated' column
playstore_data['Last Updated'] = playstore_data['Last Updated'].apply(correct_invalid_dates)

# Remove rows with invalid dates
playstore_data = playstore_data.dropna(subset=['Last Updated'])

# Convert 'Installs' to numeric by removing '+' and ',' characters and converting to int
playstore_data['Installs'] = playstore_data['Installs'].str.replace('[+,]', '', regex=True).astype(int)

# Extract month and year from 'Last Updated'
playstore_data['Last Updated Month'] = playstore_data['Last Updated'].dt.month
playstore_data['Last Updated Year'] = playstore_data['Last Updated'].dt.year

# Aggregate installs by month and year
monthly_installs = playstore_data.groupby(['Last Updated Year', 'Last Updated Month'])['Installs'].sum().reset_index()

# Line chart of installs over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_installs, x='Last Updated Month', y='Installs', hue='Last Updated Year', palette='tab10')
plt.title('Monthly Installs Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Installs')
plt.legend(title='Year')
plt.grid(True)
plt.show()

# Box plot of installs distribution by month
plt.figure(figsize=(12, 6))
sns.boxplot(data=playstore_data, x='Last Updated Month', y='Installs')
plt.title('Distribution of Installs by Month')
plt.xlabel('Month')
plt.ylabel('Number of Installs')
plt.grid(True)
plt.show()
```
