Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Cumulative Density Function (CDF):

The Cumulative Density Function (CDF) is a concept used in probability theory to describe the probability that a random variable takes on a value less than or equal to a given point. For a random variable XX, the CDF is denoted by F(x)F(x), and it is defined as:

F(x)=P(X≤x)F(x)=P(X≤x)

In other words, the CDF gives the probability that the random variable is less than or equal to a specific value xx.


Why CDF is Used?

Cumulative Probability: The CDF provides cumulative probabilities, allowing you to determine the probability that a random variable is less than or equal to a specific value.

Quantile Information: The CDF is useful for finding quantiles or percentiles. For example, you can use the CDF to determine the value below which a certain percentage of observations fall.

Comparison of Distributions: The CDF can be used to compare different probability distributions. It provides a complete picture of the distribution and is often used in statistical analysis and hypothesis testing.

Integration with PDF or PMF: The CDF is closely related to the Probability Density Function (PDF) for continuous random variables or the Probability Mass Function (PMF) for discrete random variables. The derivative of the CDF is the PDF (in the continuous case) or the PMF (in the discrete case).

```python
#Example
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


x_values = np.linspace(-3, 3, 1000)


cdf_values = norm.cdf(x_values)


plt.plot(x_values, cdf_values, label='CDF of Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Density Function of Standard Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

The normal distribution, also known as the Gaussian distribution or bell curve, is widely used to model various real-world phenomena due to its mathematical properties and the prevalence of the central limit theorem. Here are some examples of situations where the normal distribution might be used as a model:

Height of Individuals: The heights of a large population often follow a normal distribution. While there can be variations, the distribution tends to be approximately bell-shaped.

Test Scores: In educational testing, the scores of a well-designed test often follow a normal distribution. This allows for the use of statistical methods based on normality.

Measurement Errors: Errors in measurement, such as those in scientific experiments, often follow a normal distribution due to the combined effects of various small random factors.

Financial Returns: Daily stock price returns and financial asset returns are often modeled using a normal distribution, especially in quantitative finance. This assumption is a simplification but is commonly used in financial models.

IQ Scores: Intelligence quotient (IQ) scores in a population tend to be normally distributed, with the majority of people clustered around the average IQ score.




Parameters of the Normal Distribution:

The normal distribution is characterized by two parameters:

Mean (μμ):
The mean is the central value around which the distribution is centered. It represents the location of the peak of the bell curve. Shifting the mean to the left or right results in a horizontal translation of the entire distribution.

Standard Deviation (σσ):
The standard deviation measures the spread or dispersion of the distribution. A larger standard deviation leads to a wider and flatter bell curve, while a smaller standard deviation results in a narrower and taller curve.


The shape of the normal distribution is influenced by these parameters as follows:

A larger standard deviation leads to a flatter and more spread-out distribution.
A smaller standard deviation results in a taller and more concentrated distribution.

Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset
is normally distributed, what is the probability that a randomly selected observation will be greater
than 60? Use the appropriate formula and show your calculations.

```python
from scipy.stats import norm


mean = 50
std_dev = 10
specific_value = 60


z_score = (specific_value - mean) / std_dev


probability = 1 - norm.cdf(z_score)


print(f"The probability that a randomly selected observation will be greater than 60 is: {probability:.4f}")
```

Q7: Explain uniform Distribution with an example.

The Uniform Distribution is a probability distribution where all outcomes are equally likely. In other words, each possible outcome has the same probability of occurring. The probability density function (PDF) of a continuous uniform distribution is constant within a specific interval and zero outside that interval. The uniform distribution is often denoted as U(a,b)U(a,b), where aa and bb are the parameters defining the interval.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform


a = 0  
b = 1  


random_samples = uniform.rvs(loc=a, scale=(b-a), size=1000)


plt.hist(random_samples, bins=20, density=True, edgecolor='black')
plt.xlabel('Random Samples')
plt.ylabel('Probability Density')
plt.title('Uniform Distribution U(0, 1)')
plt.show()
```

Q8: What is the z score? State the importance of the z score.

Quality Control:
        In quality control processes, Z-scores are often used to assess how far a data point is from the mean and whether it falls within an acceptable range.



Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

The Central Limit Theorem is a fundamental concept in probability and statistics. It states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables approaches a normal (Gaussian) distribution, regardless of the original distribution of the variables. This holds true as long as the sample size is sufficiently large.

Significance of the Central Limit Theorem:

Normality of Sample Means:
        The CLT implies that even if the underlying population distribution is not normal, the distribution of sample means will tend to be normal for large sample sizes. This is particularly powerful because it allows for the use of normal distribution-based statistical methods in a wide range of practical scenarios.

Statistical Inference:
        The CLT is the foundation for many statistical inference methods. It justifies the use of parametric statistical tests and confidence intervals based on the assumption of normality, even when dealing with non-normally distributed populations.

Sampling Distribution Approximation:
        The CLT simplifies the study of sampling distributions. It allows researchers to approximate the sampling distribution of the sample mean, regardless of the shape of the original population distribution.

Risk Management:
        In risk management and finance, the CLT is essential. It facilitates the analysis of portfolios and financial instruments by allowing the use of normal distribution-based models, such as the calculation of value-at-risk (VaR).

Quality Control:
        In quality control processes, the CLT is applied to assess the distribution of sample means, enabling better understanding and control of production processes.

Q10: State the assumptions of the Central Limit Theorem.

The Central Limit Theorem (CLT) is a powerful statistical concept, but it relies on certain assumptions for its validity. Here are the key assumptions of the Central Limit Theorem:

Independence:
        The observations in the sample must be independent. This means that the occurrence or value of one observation should not influence the occurrence or value of another. Independence is crucial for the averaging process to work effectively.

Identically Distributed:
        The random variables in the sample must be identically distributed, meaning that they should follow the same probability distribution. This assumption ensures that the characteristics of the individual random variables are consistent.

Finite Mean and Variance:
        The random variables should have a finite mean (μμ) and a finite variance (σ2σ2). The mean and variance provide measures of central tendency and variability, and their finiteness is necessary for the convergence to a normal distribution.

Sample Size is Sufficiently Large:
        The Central Limit Theorem works best when the sample size (nn) is sufficiently large. While there is no strict rule for what constitutes "sufficiently large," a common guideline is that nn should be greater than or equal to 30. Larger sample sizes generally result in a closer approximation to a normal distribution.

No Heavy Tails:
        The population distribution should not have heavy tails or extreme skewness. Heavy-tailed distributions may not converge to a normal distribution as quickly, and the CLT may not be as effective in such cases.
